```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv('datasets/Salary_dataset.csv')

df.drop(['Unnamed: 0'], axis=1, inplace=True)

plt.scatter(df['YearsExperience'], df['Salary'])
```

```python
# SPlit it into Independent and Dependent variable
years_X, salary_Y = df['YearsExperience'].values.reshape(-1,1), df['Salary'].to_list()

# years_X
# Split into train and testing 8:2
year_X_Train = years_X[:26]
year_X_Test = years_X[-6:]

# Split into train and testing 8:2
salary_Y_Train = salary_Y[:26]
salary_Y_Test = salary_Y[-6:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(year_X_Train, salary_Y_Train)

# Make a prediction
salary_Y_pred = regr.predict(year_X_Test)

print("Coefficients: %.2f"% regr.coef_[0])
print("Mean squared error: %.2f" % mean_squared_error(salary_Y_Test, salary_Y_pred))
print("Coefficient of determination: %.2f" % r2_score(salary_Y_Test, salary_Y_pred))

# Plot outputs
plt.scatter(year_X_Test, salary_Y_Test, color="black")
plt.plot(year_X_Test, salary_Y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()

# regr.coef_
# mean_squared_error(salary_Y_Test, salary_Y_pred)
# r2_score(salary_Y_Test, salary_Y_pred)
```

```python
trainSet = pd.read_csv("./datasets/salary/train.csv")
testSet = pd.read_csv("./datasets/salary/test.csv")

# Data cleaning
trainSet.dropna(inplace=True)
testSet.dropna(inplace=True)

train_X, train_y = trainSet['x'].values.reshape(-1,1), trainSet['y'].to_list()
test_X, test_y = testSet['x'].values.reshape(-1,1), testSet['y'].to_list()

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(train_X, train_y)

# Make a prediction
y_pred = regr.predict(test_X)

print("Coefficients: %.2f"% regr.coef_[0])
print("Mean squared error: %.2f" % mean_squared_error(test_y, y_pred))
print("Coefficient of determination: %.2f" % r2_score(test_y, y_pred))

# Plot outputs
plt.scatter(train_X, train_y, color="black")
plt.plot(test_X, y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
